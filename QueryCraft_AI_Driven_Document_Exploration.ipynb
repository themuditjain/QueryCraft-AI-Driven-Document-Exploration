**Driver Code to calculate the similarity between the questions of two different question papers as well as answering the respective questions too**

```python
import docx
import openai

# Set up OpenAI API credentials
openai.api_key = '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'

# Function to compute similarity between two questions using OpenAI API
def compute_similarity(question1, question2):
    response = openai.Completion.create(
        engine="davinci",
        prompt=question1,
        max_tokens=100,
        n=1,
        stop=None,
        temperature=0
    )
    similarity = response.choices[0].text
    return similarity


# Load the Word files
doc1 = docx.Document('document1.docx')
doc2 = docx.Document('document4.docx')

# Extract the questions from both files
questions1 = [paragraph.text for paragraph in doc1.paragraphs if paragraph.text.strip()]
questions2 = [paragraph.text for paragraph in doc2.paragraphs if paragraph.text.strip()]

# Compare each question from file1 with all questions from file2
with open('output.txt', 'w') as file:
    for question1 in questions1:
        for question2 in questions2:
            similarity = compute_similarity(question1, question2)
            output = f"Similarity between '{question1}' and '{question2}': {similarity}\n"
            file.write(output)
            print(output)
```
